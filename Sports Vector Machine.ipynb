```python
# Libraries
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from svm_visualization import draw_boundary
from players import aaron_judge, jose_altuve, david_ortiz

fig, ax = plt.subplots()

# Exploring data
print('------------------------------------------------')
print(aaron_judge.columns)
print('------------------------------------------------')

# Feature description
print(aaron_judge.description.unique())
print('------------------------------------------------')

# Unique values in type feature to get a sense of how balls and strikes are recorded.
print(aaron_judge.type.unique())

aaron_judge['type'] = aaron_judge['type'].map({'S':1,'B':0})
print(aaron_judge.type.unique())

# Plotting the pitches
print(aaron_judge['plate_x'])

#dropping all nan row values from specific columns
aaron_judge = aaron_judge.dropna(subset=['plate_x','plate_z','type'])

# Plotting points
plt.scatter(x=aaron_judge.plate_x, y=aaron_judge.plate_z, c = aaron_judge.type, cmap = plt.cm.coolwarm, alpha = 0.25)
plt.title('Pitch Coordinate')
plt.xlabel('x-plate')
plt.ylabel('y-plate')


# Training and Validation set
training_set,validation_set = train_test_split(aaron_judge,random_state = 1)

# SVM classifier
classifier = SVC(kernel='rbf',gamma = 100, C=100)

# Calling .fit() method
classifier.fit(training_set[['plate_x','plate_z']],training_set[['type']])

# Drawing boundary
ax.set_ylim(-2, 2)
draw_boundary(ax, classifier)
plt.show()

# Validation score
print(classifier.score(validation_set[['plate_x', 'plate_z']], validation_set[['type']]))

# 
def svm(data_frame):
    fig, ax = plt.subplots()
    data_frame['type'] = data_frame['type'].map({'S' : 1, 'B' : 0})
    data_frame = data_frame.dropna(subset = ['plate_x', 'type', 'plate_z'])
    plt.scatter(data_frame.plate_x, data_frame.plate_z, c = data_frame.type, cmap = plt.cm.coolwarm, alpha = 0.25)
    plt.title('Pitch Coordinate')
    plt.xlabel('x-plate')
    plt.ylabel('y-plate')
    training_set, validation_set = train_test_split(data_frame, random_state = 1)
    classifier = SVC(kernel = 'rbf', gamma = 9, C= 6)
    classifier.fit(training_set[['plate_x', 'plate_z']], training_set[['type']])
    ax.set_ylim(-2, 4)
    ax.set_xlim(-3, 3)
    draw_boundary(ax, classifier)
    plt.show()
    print(classifier.score(validation_set[['plate_x', 'plate_z']], validation_set[['type']]))

# calling funtion on david_ortiz and jose_altuve
svm(david_ortiz)
svm(jose_altuve)
```
